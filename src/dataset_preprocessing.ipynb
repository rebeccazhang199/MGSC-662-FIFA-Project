```python
import pandas as pd
import numpy as np


import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation

file = "C:/Users/TYS/Documents/study/Master/MGSC662/group project/MGSC-662-FIFA-Project/data/players_22.csv"
#file = "players_22.csv"
try:
    data = pd.read_csv(file, encoding='ISO-8859-1')
except UnicodeDecodeError:
    # If 'ISO-8859-1' fails, try with 'windows-1252' encoding
    try:
        data = pd.read_csv(file, encoding='windows-1252')
    except UnicodeDecodeError as e:
        # If both encodings fail, raise the error
        raise e

# Display the first few rows of the dataframe
data.head()
```

```python
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()

# Check for duplicate rows
duplicates = data.duplicated().sum()

# Check for columns with mixed data types
mixed_types = {col: data[col].apply(type).nunique() for col in data.columns if data[col].apply(type).nunique() > 1}

# Summarize the findings
missing_values_summary = missing_values[missing_values > 0]
mixed_types_summary = {col: types for col, types in mixed_types.items() if types > 1}

(missing_values_summary, duplicates, mixed_types_summary)
```

```python
# List of columns to drop
columns_to_drop = [
     'player_url', 'long_name', 'player_face_url', 
    'club_logo_url', 'club_flag_url', 'nation_logo_url', 'nation_flag_url', 
    'Unnamed: 110'
]

# Additionally, drop columns with more than 50% missing values
columns_to_drop.extend(missing_values[missing_values > (data.shape[0] / 2)].index.tolist())

# Drop the columns
data_cleaned = data.drop(columns=columns_to_drop, errors='ignore')

# Verify the drop operation
data_cleaned.info()
```

```python
data_cleaned
```

```python
# Identify rows with numerical values in "player_positions"
rows_with_numerical_positions = data_cleaned['player_positions'].str.contains('\d', regex=True)

# Count the number of such rows
numerical_positions_count = rows_with_numerical_positions.sum()

# Drop rows with numerical values in "player_positions"
data_cleaned = data_cleaned[~rows_with_numerical_positions]

data_cleaned
```

```python
# Identify rows in the "overall" column that contain non-numeric values
rows_with_non_numeric_overall = data_cleaned['overall'].str.isnumeric() == False

# Count the number of such rows
non_numeric_overall_count = rows_with_non_numeric_overall.sum()

# Drop rows with non-numeric values in "overall"
data_cleaned = data_cleaned[~rows_with_non_numeric_overall]

# Show the number of rows identified and dropped
data_cleaned
```
